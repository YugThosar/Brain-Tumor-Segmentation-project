```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
df = pd.read_csv("videogamesales.csv")  
print(df.head()) 
df['global sales']=df['NA_Sales']+df['EU_Sales']+df['JP_Sales']+df['Other_Sales']
print(df.head()) 
grouped = df.groupby("Genre")['global sales'].sum()
print(grouped)
grouped.plot(kind="bar", title="Copies Sold", color="brown")
plt.show()

df_filtered = df[df['Name'].str.contains("Grand Theft Auto", case=False)]
print(df_filtered)
df_filtered['sum of EU and Japan']=df_filtered['EU_Sales']+df_filtered["JP_Sales"]
print(df_filtered)
na_sales_total = df_filtered['NA_Sales'].sum()
eu_sales_total = df_filtered['EU_Sales'].sum()
jp_sales_total = df_filtered['JP_Sales'].sum()
other_sales_total = df_filtered['Other_Sales'].sum()

sales_data = pd.Series(
    [na_sales_total, eu_sales_total, jp_sales_total, other_sales_total],
    index=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
)

sales_data.plot(
    kind='pie',
    autopct='%1.1f%%', 
    startangle=90,     
    colors=['lightblue', 'lightgreen', 'gold', 'coral'] 
)

plt.title('Distribution of Grand Theft Auto Sales by Region')


plt.axis('equal')

# Display the chart
plt.show()






```
